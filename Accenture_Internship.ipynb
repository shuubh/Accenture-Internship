```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/merged_data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
unique_categories = df['Category'].nunique()
print("Number of unique categories:", unique_categories)
```

```python
category_counts = df['Category'].value_counts()
most_popular_category = category_counts.idxmax()
reactions_to_most_popular_category = category_counts.max()
print("Most popular category:", most_popular_category)
print("Reactions to the most popular category:", reactions_to_most_popular_category)
```

```python
# Visualize content type counts
plt.figure(figsize=(8, 6))
df['Content Type'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Count of Content Types')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualize category counts
plt.figure(figsize=(8, 6))
df['Category'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Count of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualize sentiment scores
plt.figure(figsize=(8, 6))
plt.hist(df['Sentiment'], bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Sentiment Scores')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['month'] = df['Datetime'].dt.month
month_counts = df['month'].value_counts()
month_with_most_posts = month_counts.idxmax()
print("Month with the most posts:", month_with_most_posts)
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['time'] = df['Datetime'].dt.time
time_counts = df['time'].value_counts()
time_of_day_most_posts = time_counts.idxmax()
print("Time of the Day most posts:", time_of_day_most_posts)
```

```python
month_counts.plot(kind='bar')
plt.title('Number of Posts per Month')
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.show()
```

```python
time_counts.plot(kind='bar')
plt.title('Number of Posts per hours')
plt.xlabel('Hours')
plt.ylabel('Number of Posts')
plt.xticks(rotation=0)
plt.show()
```
